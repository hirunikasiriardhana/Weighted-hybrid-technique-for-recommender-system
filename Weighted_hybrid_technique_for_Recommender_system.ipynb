```python
#used to modify the column names, making it easier to merge with another DataFrame that uses the same column name
credits_column_renamed = credits.rename(index=str, columns={"movie_id": "id"})
#The merge function combines rows from both DataFrames where the id values match.
movies_df_merge = movies_df.merge(credits_column_renamed, on='id')
#This displays the first 5 rows of the merged DataFrame
movies_df_merge.head()
```

```python
#The code removes specific columns from movies_df_merge, creating a cleaned DataFrame movies_cleaned_df for further analysis.
movies_cleaned_df = movies_df_merge.drop(columns=['homepage', 'title_x', 'title_y', 'status','production_countries'])
movies_cleaned_df.head()
```

```python
#The info() method provides a summary of the DataFrame movies_cleaned_df, including the number of entries, column names, data types, and memory usage.
movies_cleaned_df.info()
```

```python
# Calculate all the components based on the above formula
#v: Number of votes for each movie.
#R: Average rating for each movie.
#C: Mean of all average ratings.
#m: 70th percentile of the vote counts.
v=movies_cleaned_df['vote_count']
R=movies_cleaned_df['vote_average']
C=movies_cleaned_df['vote_average'].mean()
m=movies_cleaned_df['vote_count'].quantile(0.70)
```

```python
#R is the average rating, v is the vote count,C is the mean rating, and m is the 70th percentile of vote counts.
movies_cleaned_df['weighted_average']=((R*v)+ (C*m))/(v+m)
```

```python
#To display the first few rows of movies_cleaned_df, including the newly added weighted_average column, use
movies_cleaned_df.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Using Abhishek Thakur's arguments for TF-IDF
tfv = TfidfVectorizer(min_df=3,  max_features=None,
            strip_accents='unicode', analyzer='word',token_pattern=r'\w{1,}',
            ngram_range=(1, 3),
            stop_words = 'english')

# Filling NaNs with empty string
movies_cleaned_df['overview'] = movies_cleaned_df['overview'].fillna('')
```

```python
# Fitting the TF-IDF on the 'overview' text
tfv_matrix = tfv.fit_transform(movies_cleaned_df['overview'])
```

```python
tfv_matrix.shape
```

```python
from sklearn.metrics.pairwise import sigmoid_kernel

# Compute the sigmoid kernel
sig = sigmoid_kernel(tfv_matrix, tfv_matrix)
```

```python
# Reverse mapping of indices and movie titles
indices = pd.Series(movies_cleaned_df.index, index=movies_cleaned_df['original_title']).drop_duplicates()
```

```python
indices
```

```python
def give_rec(title, sig=sig):
    # Get the index corresponding to original_title
    idx = indices[title]

    # Get the pairwsie similarity scores
    sig_scores = list(enumerate(sig[idx]))

    # Sort the movies
    sig_scores = sorted(sig_scores, key=lambda x: x[1], reverse=True)

    # Scores of the 10 most similar movies
    sig_scores = sig_scores[1:11]

    # Movie indices
    movie_indices = [i[0] for i in sig_scores]

    # Top 10 most similar movies
    return movies_cleaned_df['original_title'].iloc[movie_indices]
```

```python
# Testing our content-based recommendation system with the seminal film Spy Kids
give_rec('Spy Kids')
```
